#  Detect Outliers :

## Boxplot Visualization

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data['Annual_Premium'])
plt.title('Outlier Detection in Annual Premium')
plt.show()
```

# Insights
* Outliers are extreme values that differ significantly from normal observations.
* In insurance datasets, outliers may occur because of:
  * luxury vehicles,
  * unusually high premiums,
  * or incorrect data entries.
* Detecting outliers is important because they can:
  * distort averages,
  * affect correlations,
  * and reduce machine learning model performance.

# Handle Outliers Using IQR Method :

```python
Q1=data['Annual_Premium'].quantile(0.25)
Q3=data['Annual_Premium'].quantile(0.75)

IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR

data = data[
    (data['Annual_Premium'] >= lower_limit) &
    (data['Annual_Premium'] <= upper_limit)
]
```

## Verify Outlier Removal:

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data['Annual_Premium'])
plt.title('After Removing Outliers')
plt.show()
```

# Insights (Outliers) :
* The IQR (Interquartile Range) method effectively identifies extreme premium values without affecting normal observations.
* Removing outliers improves:
  * data consistency,
  * statistical stability,
  * and visualization clarity.
 * Outlier handling is especially important in insurance analytics because abnormal premium values can bias business insights and predictive models.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Response',y='Vintage',data=data)
plt.title('Vintage vs Insurance Claims')
plt.xlabel('Insurance Claim')
plt.ylabel('Vintage (Days)')
plt.show()
```

# Insights

* If the graph remains relatively stable, it suggests that customer duration alone does not strongly influence claim frequency.
* Fluctuating peaks may indicate specific customer segments that require closer monitoring by the insurance company.
